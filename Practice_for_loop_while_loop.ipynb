```python
#for else statement is used if all the elements are successfuly parsed through each element without interuuption then only the else part will be executed.
#If for loop is not successfully executed then we wont be able to execute the else part
l=['name','email_id','phone_no','address']
for i in l:
    print(i)
else:
    print("If for loop is going to complete itself then it will come to else")
```

```python
##While Loop
a=1
while a<6:
    print(a)
    a=a+1
    
```

```python
a=1
while a<6:
    print(a)
    if a==4:
        break
    a=a+1
    
```

```python
#continue will give control to the loop.It will not execute the next operation and will give immediate control to the while loop 
#That is why is is printing 4 infinite times
a=1
while a<6:
    print(a)
    if a==4:
        continue
    a=a+1
```

```python
#If we want to break series of 4 with conntinue keyword
a=1
while a<5:
    print(a)
    a=a+1
    if a==3:
        continue
    
```

```python
#Pass keyword can be used within while and for loop.If we do not write anything inside while loop or for loop .It will give us syntax error.
#If we donot want any error to come we will use pass keyword.
while a<5:
    pass
```

```python
n=5
for i in range(0,n): #Here range will be [0,1,2,3,4]
    for j in range(0,i+1):#Here range will be [0]
        print("*",end ="")
    print("\r")#if u dont want spaces in ur output then you can use \r otherwise use \n which will create new space lines
```

```python
#How to use dictionary with for loop
d={"a":"fsdfds","b":"abcd","c":[1,2,3,4,5],"d":(4,5,6,7),"e":"sudh"}
```

```python
#This will return me the keys only
for i in d:
    print(i)
```

```python
d["d"]
```

```python
#But if i want to extract key with value pair of  dictionary using for loop
for i in d:
    print(i,d[i])
```

```python
d.items()
```

```python
for i in d.items():
    print(i)
```
